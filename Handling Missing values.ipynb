```python
##Imputation missing values
## Mean value imputataion
```

```python
sns.displot(df['age'])
```

```python
sns.histplot(df['age'],kde = True)
```

```python
df['Age_median'] = df['age'].fillna(df['age'].mean()) # replace NA value with median
```

```python
df[['Age_median','age']]
```

```python
df['Age_mean'] = df['age'].fillna(df['age'].mean())) # replace NA value with median
```

```python
df[['Age_mean','age']]
```

```python
## Mean Imputation works well when we have normally distribited data/gaussian
```

```python
## 2.Median value Imputation. If we have OUTLIERS in data set
```

```python
df['Age_median'] = df['age'].fillna(df['age'].mean()) 
```

```python
## Mode Imputataion technique: categorical values
```

```python
df[df['embarked'].isnull()]
```

```python
df['embarked'].unique()
```

```python
df[df['embarked'].notna()]
```

```python
mode_value= df[df['embarked'].notna()]['embarked'].mode()[0]
```

```python
mode_value
```

```python
df['embarked_mode'] = df['embarked'].fillna(mode_value)
```

```python
df[['embarked_mode', 'embarked']]
```

```python
df['embarked_mode'].isnull().sum()
```
